```python
X= n_d.drop('target',axis=1)
y= n_d['target']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV

X_trian,X_test,y_train,y_test = train_test_split(X,y, test_size= 0.3,shuffle=False)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

pipeline = Pipeline([('scaler',StandardScaler()),( 'clf',RandomForestClassifier())])

params = {
    'clf__max_depth' : [6,8,10],
    'clf__min_samples_leaf' : [15,20,25]
}

grid_model = GridSearchCV(estimator=pipeline , param_grid= params, cv=3, refit=True, scoring='accuracy')
grid_model.fit(X_trian,y_train)
```

```python
grid_model.best_params_
```

```python
y_pred=grid_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_pred,y_test))
```
