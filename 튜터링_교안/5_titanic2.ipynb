## 1-2 결측치 확인 (대체 및 제거)
 - Age -> Age 평균값으로 대체합니다.
 - Embarked -> 결측치가 있는 행 제거합니다.
 - Cabin -> 컬럼 제거합니다.

### Age

```python
# 결측치 확인
df.isna().sum()
```

```python
# Age 평균값 확인
df['Age'].mean()
```

```python
# 결측치 평균값으로 채우는 코드
df['Age'].fillna(df['Age'].mean(), inplace=True)

# 결측치 잘 채워졌는지 다시 확인
df.isna().sum()
```

### Embarked 

```python
# Embarked 결측치가 있는 데이터 확인
df[df['Embarked'].isna()]
```

```python
# Embarked 결측치가 있는 데이터 제거 후 데이터프레임 추출
df = df[~(df['Embarked'].isna())]  # df.dropna(subset='Embarked', inplace=True)
df
```

```python
# 인덱스 초기화
df.reset_index(drop=True, inplace=True)
df
```

## 1-3 불필요 컬럼 삭제
 - Name
 - Ticket
 - Cabin
 * 3개의 컬럼을 제거합니다.

```python
df.drop(columns=['Name', 'Ticket', 'Cabin'], axis=1, inplace=True, errors='ignore')
df
```

```python
# 최종 결측치 확인
df.isna().sum()
```

## 1-4 원핫인코딩
 - Sex 컬럼 ->  male : 0 female : 1 로 변경합니다.

```python
df['Sex'] = df['Sex'].map(lambda x: 0 if x == 'male' else 1)
df
```

# 1-5 더미화
 - Embarked 컬럼을 더미변수(범주형 변수를 연속형 변수로 변경)로 변경합니다.

```python
# get_dummies() 함수를 사용하여 'Embarked' 열을 더미 변수로 변환
df_dummy = pd.get_dummies(df['Embarked'], prefix='Embarked')

# 원본 데이터프레임과 더미 변수 데이터프레임을 합침
df = pd.concat([df, df_dummy], axis=1)

df.drop(columns=['Embarked'], axis=1, inplace=True, errors='ignore')
df
```

# train_test_split 진행
- x, y 값을 지정하고 7:3으로 진행합니다. 
    * Train Data / Test Data 1,0 비율은 동등하게 나눕니다.
- y : Survived

```python
# 원래 test 파일 따로 써야하지만 test파일이 존재하지 않으니 해당 파일을 train_test_split 해줍니다.
from sklearn.model_selection import train_test_split

X = df.drop(columns=['Survived'], axis=1)
y = df[['Survived']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=1)
```

```python
# 모델 성능평가
from sklearn.metrics import confusion_matrix

cf_matrix = confusion_matrix(y_test, y_pred)
print(cf_matrix)
```

- 모델성능평가2 : precision_score, recall_score, accuracy_score, f1_score를 프린트해봅니다.

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

print("Accuracy : %.3f" % accuracy_score(y_test, y_pred))
print("Precision : %.3f" % precision_score(y_test, y_pred))
print("Recall : %.3f" % recall_score(y_test, y_pred))
print("F1 : %.3f" % f1_score(y_test, y_pred))
```

# 기본 Modeling 함수

```python
def model_basic(X_train, y_train, X_test, y_test):
    models = [
        LogisticRegression(),
        DecisionTreeClassifier(),
        SVC(),
        RandomForestClassifier(),
        XGBClassifier(),
        LGBMClassifier()
    ]

    rdict={'model':[],'accuracy':[],'precision':[],'recall':[],'f1_score':[]}

    
    for clf in models:
        clf = clf.fit(X_train, y_train)
        pred = clf.predict(X_test)
        results = (round(accuracy_score(y_test, pred),2),
                    round(precision_score(y_test, pred),2),
                    round(recall_score(y_test, pred),2),
                    round(f1_score(y_test, pred),2))
        rdict['model'].append(clf); 
        rdict['accuracy'].append(results[0])
        rdict['precision'].append(results[1])
        rdict['recall'].append(results[2])
        rdict['f1_score'].append(results[3])

        # print(results)

    rdf = pd.DataFrame(data=rdict)
    return rdf 
```

```python
model_basic(X_train, y_train, X_test, y_test)
```
